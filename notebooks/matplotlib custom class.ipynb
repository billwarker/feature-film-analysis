```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
x = np.linspace(0,10,20)
y = np.random.randint(0,100, size=len(x))
```

```python
plt.figure(figsize=(16,10))
plt.scatter(x,y)
plt.plot(x,y)
plt.title('my plot', fontsize=24)
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
```

understanding \*args and \**kwargs

```python
# # first with *args
# >>> args = ("two", 3,5)
# >>> test_args_kwargs(*args)
# arg1: two
# arg2: 3
# arg3: 5

# # now with **kwargs:
# >>> kwargs = {"arg3": 3, "arg2": "two","arg1":5}
# >>> test_args_kwargs(**kwargs)
# arg1: 5
# arg2: two
# arg3: 3
```

args passes in unnamed arguments, kwargs is for named args. args and kwargs are just the conventional names for doing this, what matters is * for unnamed and ** for named.

\*var and \**keywordvar would also work

```python
### custom class

class CustomPlot(matplotlib.figure.Figure):
    def __init__(self, *args, **kwargs):
        """
        class with my preferred plotting presets
        """
        super().__init__(figsize=(16,10))
```

```python
plt.figure(FigureClass=CustomPlot)
plt.plot(x,y)
```
