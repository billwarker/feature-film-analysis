```python
engine = create_engine(os.environ.get("DATABASE_URL"))
```

## Ratings over time

```python
films = pd.read_sql_query(scripts.read("film_view1.sql"), engine)
```

```python
films = films.set_index(pd.DatetimeIndex(films.released))
films = films[films.index.year > 2009]
```

```python
rated_dummies = pd.get_dummies(films.rated)
films_rated = pd.concat([films, rated_dummies], axis=1)
```

```python
films_rated.resample('Y').count()
rating_cols = films_rated.columns[14:]
yearly_ratings = films_rated.resample('Y').count()[rating_cols.values]
```
